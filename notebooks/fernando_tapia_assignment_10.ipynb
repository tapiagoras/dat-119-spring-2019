```python
'''

 Fernando Tapia Tinoco 
 4/15/2019
 Python 1 - DAT-119 - Spring 2019
 Assignment # 10
     Program that converts a text file into Pig Latin

'''
import os
import string


#function 
#This function will validate if the file exist to avoid run any code if file does not exist

def check_file_exists(name_of_file):
    
    # exists will be True if the file is there, False if not 
    exists = os.path.isfile(name_of_file) #Coral Sheldon snippet
    if (exists):
        my_data_object = open(name_of_file, "r+")
    else:
        my_data_object = ""
    
    return my_data_object

#This function receive a line of a file and return the same line translated to pig_latin
#This is the simple version without vowel or double consonants validation.
def convert_to_pig_latin(poem_lines):
    
    poem_lines = poem_lines.rstrip('\n')#get rid of the extra line to avoid a translation of 
                                        #an empty espace
    words = poem_lines.split()#list of words
    #print (words)
    pig_line = []
    for word in words:
        clean_word = []#this list will hold every word without punctuation
        for character in word:
            if character not in string.punctuation:#this will check every letter to create a 
                                                   #word without punctuation.
                clean_word.append(character)
        clean_word = "".join(clean_word)
        
        #print(word)
        #print(clean_word)
        #this will convert the word without symbols to a pig_latin word
        if clean_word:
            pig_word = clean_word[1:len(clean_word)] + clean_word[0] + "ay"
        else:
            pig_word = clean_word + "ay"
        
        if clean_word.istitle():#if the world was capitalize then the latin_pig word 
                                #should be capitalized as well
            pig_word = pig_word.capitalize()
        #print (pig_word)
        #this will replace the new word in the position of the clean word to recover punctuation
        pig_word_with_symbols = word.replace(clean_word, pig_word)
        #print(pig_word_with_symbols)
        #print("____")
        pig_line.append(pig_word_with_symbols)
                          
    pig_line = " ".join(pig_line)    
    return pig_line

#this function is needed for the second version of the translation
#if will return true if the character is vowel
def isvowel(character):
    vowels=["a","e","i","o", "u"]
    for index in range (len(vowels)):
        if character.lower() == vowels[index]:
            return True
    return False
       
#second version of the pig translation
def convert_to_pig_latin_v2(poem_lines):
    
    poem_lines = poem_lines.rstrip('\n')
    words = poem_lines.split()
    #print (words)
    pig_line = []
    for word in words:
        
        clean_word = []
        for character in word:
            if character not in string.punctuation:
                clean_word.append(character)
        clean_word = "".join(clean_word)
        
        #print(word)
        #print(clean_word) 
        #this will convert the word without symbols to a pig_latin word
        #but is going to check also for vowelss and double consonants
        if clean_word:
            if isvowel(clean_word[0]):
                pig_word = clean_word + "ay"
            elif not isvowel(clean_word[1]):
                pig_word = clean_word[2:len(clean_word)] + clean_word[0:2] + "ay"
            else:
                pig_word = clean_word[1:len(clean_word)] + clean_word[0] + "ay"       
        else:
            pig_word = clean_word + "ay"
        
        if clean_word.istitle():
            pig_word = pig_word.capitalize()
        #print (pig_word)
       
        pig_word_with_symbols = word.replace(clean_word, pig_word)
        #print(pig_word_with_symbols)
        #print("____")
        pig_line.append(pig_word_with_symbols)
                          
    pig_line = " ".join(pig_line)    
    return pig_line
        
def main():
    
    poem_object = check_file_exists('prejudice.txt')
    
    
    if  poem_object:
        poem_list = poem_object.readlines()
        poem_object.close()
        #create new object to storage the new poem
        poem_latin_object = open("prejudice_latin.txt", "w")
    
        for index in range(len(poem_list)):
            #uncomment next line and comment following line for original version 
            #pig_poem_line = convert_to_pig_latin(poem_list[index])
            
            #This will call version 2
            pig_poem_line = convert_to_pig_latin_v2(poem_list[index])
            
            poem_latin_object.write(pig_poem_line+"\n")
        
        poem_latin_object.close()
        print("Translation Complete")
    else:
        print("File does not exist.")

if __name__ == "__main__":
	main()
```
