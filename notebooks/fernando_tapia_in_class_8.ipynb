```python
'''

 Fernando Tapia Tinoco 
 3/27/2019
 Python 1 - DAT-119 - Spring 2019
 In-class Assignment # 8
     Program that handle a TO-DO list.

'''
#function that display a list content
# message: print the title previous to the list
# selected_list: list to be displayed
# wait_for_user: add the enter option at the end to make a pause
def view_list(message,selected_list,wait_for_user):
    print(message)
    if len(selected_list):
        #loop to display the list items
        for item in range(len(selected_list)):
            print(item+1,") ",selected_list[item],sep="")
       
    else:
        print("There is nothing to display.")
        
    # pause until user hits enter
    if(wait_for_user):
        print("")
        input("Hit enter when finished viewing ")
        print("")
#function that add an item to the list    
def add_todo_list(todo_list):
    new_item = input("What would you like to add? ")
    if new_item not in todo_list:
        todo_list.append(new_item)
    else:
        print(new_item, "is already in the list.")

#function to remove an item from the list
def complete_task(todo_list,completed_list):
    #print the items availables on todo_list
    view_list("Choose an item to mark completed. ",todo_list,False)
    if (len(todo_list)>0):#validate that the todo list has at least one item
        user_selection = validate_selection(todo_list) #validates the user selection
        #the following line storage the value of the user_selection for future usage
        user_selection_item = todo_list[user_selection-1] 
        todo_list.remove(user_selection_item)
        completed_list.append(user_selection_item)
    return completed_list
    
#Function to validate the user input based on the number of available options
def validate_selection(list_to_evaluate):
    user_selection = 0
    input_wrong = True
    print("")
    while user_selection < 1 or user_selection > len(list_to_evaluate):
        while input_wrong:
            user_selection = input("Please choose one of the options above: ")
            if user_selection.isdigit():
                if(int(user_selection) > len(list_to_evaluate)):
                    input_wrong = True
                else:
                    user_selection = int (user_selection)
                    input_wrong = False
            else:
                input_wrong = True
    print("")
    return user_selection

        
def main():
    
    #define lists
    menu_list = ("View the items on your todo list","View the items you have finished",
             "Add an item to the todo list","Mark a todo list item as completed",
            "Exit the Todo List Application (your list will not be saved)")
    todo_list = []
    completed_list = []
   
    print("Welcome to the Todo List Application.")
    
    app_is_running = True
    
    
    #Loop to keep the app running
    while (app_is_running):
        view_list ("",menu_list,False)
        user_selection = validate_selection(menu_list)
        if user_selection == 1:
            view_list ("This is your todo list: ",todo_list,True)
        elif user_selection == 2:
            view_list ("This is your list of completed tasks: ",completed_list,True)
        elif user_selection == 3:
            add_todo_list(todo_list)
        elif user_selection == 4:
            completed_list = complete_task(todo_list,completed_list)
        if user_selection == 5:
            app_is_running = False
            print("Thank you for using the Todo List Application.")

if __name__ == "__main__":
	main()
```
