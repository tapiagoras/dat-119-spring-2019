```python
import random
#print(states_capital_dic.keys())


def random_state():
    return random.choice(list(states_capital_dic.keys()))



def main():
    play = True #artifice to avoid an input outside the loop
    #Loop for playing multiple times
    print("Welcome to Guess a Capital Game (?)")#reusing code for all assignment
    print("¯\_(ツ)_/¯")
    print("")
    correct = 0
    incorrect = 0  
    
    while (play):
        
        state = random_state()
        
        print (states_capital_dic[state]) #for evaluation purpose
        human = input("The capital of "+ state +" is : ")
        if (states_capital_dic[state].lower() == human.lower()):
            #Evaluation to play again
            print("")
            print ("Yes, the capital of " + state + " is "+ states_capital_dic[state] + " Great Job")
            correct = correct + 1
            
            
        else:
            print("")
            print ("Sorry, the capital of " + state + " is "+ states_capital_dic[state])
            incorrect = incorrect + 1
        
        print ("Correct :", correct)
        print ("Incorrect :", incorrect)
        
        play = input("DO you want to play again (Y/N): ")
        if ("y" in play or "Y" in play):
            play = True
        else :
            print("See you next time")
            play = False
    

if __name__ == "__main__":
	main()
```
